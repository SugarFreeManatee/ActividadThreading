```python
from random import randint
from better_profanity import profanity
import twitterscraper as ts
import datetime as dt
import pandas as pd
import re

# profanity.load_censor_words()
```

```python
# Get tweets from parody Kim Jong-Un Accounte
kim_tweets = ts.query_tweets_from_user("_Kim_Jongun", 1000)
```

```python
# Get tweets from official Trump account realDonaldTrump 
trump_tweets = ts.query_tweets_from_user("realDonaldTrump", 500)
```

```python
trump_tweets2 = ts.query_tweets_from_user("realDonaldTrFan", limit=3000)
```

```python
trump_tweets3 = ts.query_tweets_from_user("RealDonalDrumpf", 1000)
```

```python
def filter_tweets(tweet_array, filter_words=[], ditch_words=[]):
    new_tweets = []
    for tweet in tweet_array:
        text = tweet.text.lower()
        
        # Remove tweets with unwanted words
        cont = False
        for word in ditch_words:
            if word.lower() in text:
                cont=True
                break
        if cont:
            continue

        # Keep tweets with wanted words
        for word in filter_words:
            if word.lower() in text:
                new_tweets.append(tweet)
                break
    return new_tweets

def switch_and_censor(tweet_array):
    trump_aliases = ["Donald Trump", "DonaldTrump", "Donald", "Trump"]
    kim_aliases = ["Kim Jung Un", "Kim Jung-Un", "Kim Jong-Un", "Kim-Jong-Un", "Kim-Jong Un", "Kim Jong Un", "Kim"]
    for tweet in tweet_array:
        tweet_text = tweet.text
        print("BEFORE:", tweet_text)
        for pattern in trump_aliases:
            tweet.text = re.sub(pattern, "Trumpzini", tweet_text, flags=re.I)
        for pattern in kim_aliases:
            tweet_text = re.sub(pattern, "Dr. Pin Tong-Un", tweet_text, flags=re.I)
        # tweet_text = profanity.censor(tweet_text)
        # tweet.text = temp
        print("AFTER: ", tweet_text) # CHECK IF IT BEHAVES PROPERLY

def dump_tweets(tweet_array, output_path="output.csv"):
    with open(output_path, "w", encoding='utf-8') as f:
        f.write("value;text\n")
        for tweet in tweet_array:
            username = tweet.screen_name
            value = randint(0,10)
            tweet_text = tweet.text.replace("\n"," ").encode('utf-8',errors='ignore').decode('utf-8')
            f.write(f"{username},{value},{tweet_text}\n")
```

```python
catch_words = ['kim', 'korea', 'trump', 'donald']
ditch_words = ['kim kardashian']
total_trump_tweets = trump_tweets + trump_tweets2 + trump_tweets3

total_trump_tweets = filter_tweets(total_trump_tweets, catch_words, ditch_words)
kim_tweets = filter_tweets(kim_tweets, catch_words + ['united states', 'usa'], ditch_words)

switch_and_censor(total_trump_tweets)
switch_and_censor(kim_tweets)

dump_tweets(total_trump_tweets, "trump_tweets.csv")
dump_tweets(kim_tweets, "kim_tweets.csv")
```
